## 1. What is Generative AI?

**Generative AI** refers to a class of artificial intelligence systems that can **create new content** — text, images, audio, video, code, and more — rather than simply classifying or predicting from existing data.

At its core, a generative model learns the **underlying patterns and distribution** of training data and uses that knowledge to generate new, plausible samples.

### Key Characteristics
| Characteristic | Description |
|---|---|
| **Creative output** | Produces new content, not just labels or scores |
| **Foundation models** | Trained on massive datasets, fine-tuned for tasks |
| **Multi-modal** | Can work across text, images, audio, code |
| **Prompt-driven** | Controlled via natural language instructions |
| **Emergent abilities** | Capabilities that arise from scale, not explicit programming |

## 2. Traditional AI vs Generative AI — Deep Comparison

| Dimension | Traditional AI | Generative AI |
|---|---|---|
| **Goal** | Classify, predict, detect | Create, generate, synthesize |
| **Output** | Structured (label, number) | Unstructured (text, image, audio) |
| **Training data** | Labeled datasets | Massive unlabeled corpora |
| **Task specificity** | One model, one task | One model, many tasks |
| **User interface** | API / dashboard | Natural language (prompt) |
| **Interpretability** | Often explainable | Often a black box |
| **Compute** | Moderate | Very high (training) |
| **Examples** | ResNet, XGBoost, SVM | GPT-4, DALL-E, Stable Diffusion |

### The Paradigm Shift

Traditional AI was **task-specific**: you trained a model to detect tumours, or to classify emails, or to predict churn — each a separate pipeline.

Generative AI introduced **foundation models**: a single massive model trained on broad data that can be prompted to perform hundreds of different tasks without retraining.

````python
# ── GenAI Timeline: Key Milestones ──

import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

milestones = [
    (2017, 'Transformer\nArchitecture', '#2E75B6'),
    (2018, 'BERT &\nGPT-1', '#1F4E79'),
    (2020, 'GPT-3\n(175B params)', '#1B5E3B'),
    (2021, 'DALL-E &\nCodices', '#7B3F00'),
    (2022, 'ChatGPT &\nStable Diffusion', '#C00000'),
    (2023, 'GPT-4, Gemini,\nLlama, Claude', '#9B2D92'),
    (2024, 'Multimodal &\nAgentic AI', '#FF6600'),
]

fig, ax = plt.subplots(figsize=(14, 4))
ax.set_xlim(2016, 2025)
ax.set_ylim(-1.5, 2.5)
ax.axis('off')
ax.set_title('Key Milestones in Generative AI', fontsize=14, fontweight='bold')

ax.axhline(y=0.5, xmin=0.02, xmax=0.98, color='#CCCCCC', linewidth=2, zorder=1)

for i, (year, label, color) in enumerate(milestones):
    y_text = 1.6 if i % 2 == 0 else -0.8
    y_arrow_end = 0.65 if i % 2 == 0 else 0.35
    ax.plot(year, 0.5, 'o', markersize=14, color=color, zorder=3)
    ax.plot(year, 0.5, 'o', markersize=10, color='white', zorder=4)
    ax.plot(year, 0.5, 'o', markersize=6, color=color, zorder=5)
    ax.annotate('', xy=(year, y_arrow_end), xytext=(year, y_text - (0.3 if i%2==0 else -0.3)),
                arrowprops=dict(arrowstyle='->', color=color, lw=1.5))
    ax.text(year, y_text, f'{year}\n{label}', ha='center', va='center',
            fontsize=8.5, fontweight='bold', color=color,
            bbox=dict(boxstyle='round,pad=0.3', fc='white', ec=color, lw=1.2))

plt.tight_layout()
plt.show()
print('✅ Timeline rendered')
````

## 3. How Generative AI Works (Conceptual)

### The Core Idea: Next Token Prediction

Most text-based GenAI models (LLMs) work by **predicting the most likely next token** given all previous tokens. A "token" is roughly a word or word-fragment.

```
Prompt:  "The capital of France is"
Model:   P("Paris" | "The capital of France is") = 0.97
Output:  "Paris"
```

### Training in Two Stages

1. **Pre-training**: The model reads billions of documents and learns language patterns, world knowledge, and reasoning — purely by predicting the next token.

2. **Fine-tuning / RLHF**: The model is then refined using human feedback to be helpful, harmless, and honest — transforming a language predictor into a useful assistant.

### Image Generation (Diffusion Models)

Models like DALL-E and Stable Diffusion work differently — they learn to **reverse a noising process**:
- Training: gradually add noise to images until they become pure noise
- Inference: start from noise and iteratively denoise, guided by a text prompt

> 💡 **Key insight**: GenAI models don't "look up" answers — they generate probabilistic continuations based on learned patterns.

## 4. Major GenAI Platforms

### Text / Language Models
